```python
import pandas as pd
from pycaret.classification import *

# ===== 1. Load & clean =====
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].fillna(0, inplace=True)
df_model = df.drop("customerID", axis=1)

print("Shape:", df_model.shape)
print(df_model["Churn"].value_counts())

# ===== 2. PyCaret setup (GPU if available) =====
clf_setup = setup(
    data=df_model,
    target="Churn",
    session_id=123,
    normalize=True,
    transformation=True,
    low_variance_threshold=None,
    remove_multicollinearity=True,
    multicollinearity_threshold=0.9,
    fix_imbalance=True,
    train_size=0.7,
    fold=5,          # lower folds -> faster
    use_gpu=True,    # let PyCaret / LightGBM use GPU when it can
    html=False,
    verbose=True
)

# ===== 3. Compare models (no include filter) =====
best_model = compare_models(sort="AUC")
print("\nBest model:\n", best_model)

# ===== 4. Create & tune LightGBM only =====
# LightGBM is available in your log and can run on GPU internally.[file:82][70][77]
lgbm = create_model("lightgbm")

# If warnings are crazy, you can reduce complexity & speed it up:
tuned_lgbm = tune_model(
    lgbm,
    optimize="AUC",
    n_iter=20,               # fewer iterations = faster
    early_stopping=True
)

print("\nTuned model:\n", tuned_lgbm)

# ===== 5. Evaluate =====
plot_model(tuned_lgbm, plot="auc")
plot_model(tuned_lgbm, plot="confusion_matrix")
plot_model(tuned_lgbm, plot="feature")

# ===== 6. Predict & save =====
test_predictions = predict_model(tuned_lgbm)
print("\nTest predictions (head):")
print(test_predictions.head())

final_model = finalize_model(tuned_lgbm)
save_model(final_model, "churn_lgbm_gpu")
```
